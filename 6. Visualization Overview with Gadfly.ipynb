```julia
import Gadfly
import DataFrames
import RDatasets
```

```julia
iris=RDatasets.dataset("datasets","iris");
head(iris,2)
```

```julia
Gadfly.set_default_plot_size(20cm, 10cm)
Gadfly.plot(iris, x=:SepalLength,y=:SepalWidth,color=:Species,Gadfly.Geom.point)
```

```julia
Gadfly.plot(iris,x=:Species,y=:PetalWidth,color=:Species,Gadfly.Geom.boxplot)
```

```julia
Gadfly.plot(iris,x=:PetalWidth,color=:Species,Gadfly.Geom.density)
```

```julia
Gadfly.plot(iris,x=:PetalWidth,color=:Species,xgroup=:Species,Gadfly.Geom.subplot_grid(Gadfly.Geom.histogram))
```

```julia
using DimensionalityReduction
iris=RDatasets.dataset("datasets","iris");
irissp = convert(Array,DataArray(iris[:,1:4]));
```

```julia
pcares=pca(irissp);
```

```julia
pcares.
```

```julia
Gadfly.set_default_plot_size(15cm, 10cm)
scores=pcares.scores[:,1:2];
plot(x=scores[:,1],y=scores[:,2], color=iris[:,5], Geom.point)
```
