```julia
import Gadfly
import DataFrames
import RDatasets
```

```julia
iris=RDatasets.dataset("datasets","iris");
head(iris,2)
```

```julia
Gadfly.plot(iris, x=:PetalWidth,y=:SepalLength,color=:Species,Gadfly.Geom.point)
```

```julia
Gadfly.plot(iris,x=:Species,y=:PetalWidth,color=:Species,Gadfly.Geom.boxplot)
```

```julia
Gadfly.plot(iris,x=:PetalWidth,color=:Species,Gadfly.Geom.density)
```

```julia
Gadfly.plot(iris,x=:PetalWidth,color=:Species,xgroup=:Species,Gadfly.Geom.subplot_grid(Gadfly.Geom.histogram))
```

```julia
econ = RDatasets.dataset("Zelig", "macro")
head(econ)
```

```julia
using Gadfly
Gadfly.set_default_plot_size(30cm, 10cm)
Gadfly.plot(econ, x=:Year, y=:Unem, xgroup=:Country,color=:Country, Gadfly.Geom.subplot_grid(Gadfly.Geom.point,Geom.smooth))
```

```julia
Gadfly.set_default_plot_size(20cm, 20cm/golden)
Gadfly.plot(econ, x=:Year, y=:Country, color=:GDP, Gadfly.Geom.rectbin)
```

```julia
using DimensionalityReduction
iris=RDatasets.dataset("datasets","iris");
```

```julia
irissp = convert(Array,DataArray(iris[:,1:4]));
```

```julia
pcares=pca(irissp);
```

```julia
Gadfly.set_default_plot_size(15cm, 10cm)
scores=pcares.scores[:,1:2];
plot(x=scores[:,1],y=scores[:,2], color=iris[:,5], Geom.point)
```
