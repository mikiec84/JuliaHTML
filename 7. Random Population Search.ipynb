```julia
using Gadfly
import Winston
using Distributions
using DataFrames
```

```julia
x=[-100:0.1:100]
f(x)=cos(x)./(1+0.01*x.^2)
Winston.plot(x,f(x))
```

```julia
pts=[-100:0.1:100]
f(x)=cos(x)./(1+0.01*x.^2)
Gadfly.plot(x=pts,y=f(pts),Gadfly.Geom.point)
```

```julia
d=convert(DataFrame,[pts f(pts)])
Gadfly.plot(Gadfly.layer(d,x=:x1,y=:x2,Gadfly.Geom.point),Gadfly.layer(d,x=:x1,y=:x2,Gadfly.Geom.line))
```

```julia
perf=f(rand(Distributions.Uniform(-100,100),10))
```

```julia
order=sortperm(perf)
```

```julia
@time begin
psize=10000
pertRate=0.1
population=rand(Distributions.Uniform(-100,100),psize)
fitness=f(population)
order=sortperm(fitness)
for i=1:1000
    [population[order[j]]+=pertRate*rand(Distributions.Normal(0,(psize+1-j))) for j=1:(psize-1)]
    fitness=f(population)
    order=sortperm(fitness)
  #  print(fitness[order[psize]],"\n")
end
end
(fitness[order[psize]],population[order[psize]])
```

```julia
if nprocs()==1
    addprocs(7)
end

@everywhere using Distributions
@everywhere f(x)=cos(x)./(1+0.01*x.^2)

@time begin
    @everywhere psize=10000
    @everywhere pertRate=0.1
    @everywhere population=rand(Distributions.Uniform(-100,100),psize)
    @everywhere fitness=f(population)
    @everywhere order=sortperm(fitness)
    for i=1:1000
        @parallel [population[order[j]]+=pertRate*rand(Distributions.Normal(0,(psize+1-j))) for j=1:(psize-1)]
        fitness=f(population)
        order=sortperm(fitness)
    end
end
(fitness[order[psize]],population[order[psize]])
```
